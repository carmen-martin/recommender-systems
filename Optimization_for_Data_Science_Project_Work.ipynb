## Data Cleaning

```python
true_series = [df.book_id.value_counts() > df.book_id.value_counts().quantile(0.90)]
```

```python
print(df.shape)
df = df[df.book_id < df.book_id.value_counts().quantile(0.9)]
print(df.shape)
df = df[df.user_id < df.user_id.value_counts().quantile(0.25)]
print(df.shape)
```

## Data Pivoting

```python
df = pd.pivot_table(df, columns="book_id", index="user_id", values="rating")
df.head(100)
```
